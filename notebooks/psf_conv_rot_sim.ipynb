```python
import numpy as np

import matplotlib.pyplot as plt

import scipy
```

```python
img_star = np.zeros((30,30))

img_star[15][15] = 255
```

```python
plt.imshow(img_star)
```

```python
psf = np.zeros((5,5))

psf[2][2] = 1
psf[2][3] = 1
psf[2][4] = 1
```

```python
plt.imshow(psf)
```

```python
psf_img = scipy.signal.convolve2d(img_star, psf)
```

```python
plt.imshow(psf_img)
```

```python
# 90 degree rotation
psf_90 = psf.T
```

```python
psf_img_rot = scipy.signal.convolve2d(img_star, psf_90)
```

```python
plt.imshow(psf_img_rot)
```
